```python
# Append all the different datasets to one dataset for final use
combination_data = petroleum_data.copy()
```

```python
combination_data = combination_data.append(renewables_data)
combination_data = combination_data.append(gas_data)
combination_data = combination_data.append(waste_data)
combination_data = combination_data.append(nucleair_data)
combination_data = combination_data.append(solid_fuels_data)
```

```python
combination_data.head()
```

```python
# Delete the unnecessary columns
combination_data = combination_data[["TIME","GEO","PRODUCT","Value"]]
```

```python
combination_data.head()
```

```python
# Delete unnecessary rows Values
combination_data= combination_data[combination_data.GEO != "European Union (current composition)"]
combination_data= combination_data[combination_data.GEO != "Euro area (19 countries)"]
```

```python
# Change Value outputs into a python readable way
combination_data['Value'] = combination_data.Value.apply(lambda x: x.replace(',',''))
combination_data['Value'] = combination_data.Value.apply(lambda x: x.replace(':','0'))
combination_data["Value"] = combination_data["Value"].astype(float)
```

```python
# Rename some of the countries so that it is compatible with R shiny.
combination_data["GEO"].unique()
#combination_data["GEO"] = combination_data["GEO"].drop(["European Union (current composition)","Euro area (19 countries)"])
combination_data["GEO"] = combination_data["GEO"].replace(["United Kingdom",'Germany (until 1990 former territory of the FRG)','Kosovo (under United Nations Security Council Resolution 1244/99)',"Czechia",'Former Yugoslav Republic of Macedonia, the'],["UK","Germany","Kosovo","Czech Republic","Macedonia"])
```

```python
# Merge population with the combined data to get the full needed dataset
energy_data = pd.merge(combination_data, population_data, on=["TIME", "GEO"])
```
